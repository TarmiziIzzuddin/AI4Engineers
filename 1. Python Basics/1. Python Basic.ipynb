### 2. Loops

### 'for'... loops

```python
utilities = ['pen','marker','shapener']
```

```python
# for loop while considering length of index
for i in range(len(utilities)):
    print (utilities[i].upper())
```

```python
# for loop while considering items in a list
for util in utilities:
    print(util.upper())
```

```python
# for loop for dictionary
animals = {'herbivor':'giraffe','carnivore':'lion','omnivor':'hen'}
for key,value in animals.items():
    print(key,value)
```

```python
# for loop using enumerate when access list's index

for index,utilities in enumerate(utilities):
    print(index,utilities)
```

```python
# using for loop to find elements
utilities = ['pen','marker','shapener']
for util in utilities:
    if util=='marker':
        print("Found the 'marker'!")
        break
    else:
        print("Not found the 'marker'")
```

### 'while'....loops

```python
count = 0
while count<10:
    print(count,' number of prints')
    count += 1
```

### 3. Effect of loops in lists

```python
# list can only be append for new list when involve loops

# area of a square 
lengths = [1,2,3,4,5]
areaSquare = []
for leng in lengths:
    areaSquare.append(leng**2)
```

```python
areaSquare
```

```python
# simplify for
lengths = [1,2,3,4,5]
areaSquare = [leng**2 for leng in lengths]
print(areaSquare)
```

```python
#area of even square
lengths = [1,2,3,4,5]
areaEvenSquare=[]

for leng in lengths:
    if (leng%2 == 0):
        areaEvenSquare.append(leng**2)
    else:
        areaEvenSquare.append(leng)

print(areaEvenSquare)
```

```python
# simplify condition
lengths = [1,2,3,4,5]
areaEvenSq = [leng**2 if leng%2==0 else leng for leng in lengths]
print(areaEvenSq)
```

```python
# flatten matrix
mat1 = [[10,11],[12,13]]
```

```python
newMat = []
for row in mat1:
    for inner in row:
        newMat.append(inner)
print(newMat)
```

```python
#simplify condition
newM = [inner for row in mat1 for inner in row]
print(newM)
```
